```python
import yaml
import os

credentials = yaml.load(open(os.path.expanduser('googleapi.yml')))
# oauth2token = 
```

```python
import json
import requests

api_key = credentials['api_key']

url = "https://www.googleapis.com/qpxExpress/v1/trips/search?key=" + api_key
headers = {'content-type': 'application/json'}
```

```python
import datetime as dt
```

```python
def qfx(datelist, orig, dest):
    results = []
    
    for i in range(len(datelist)):
        params = {
            "request": {
            "slice": [
          {
            "origin": orig,
            "alliance": "ONEWORLD",
            "destination": dest,
            "date": "{}".format(datelist[i]),
            "permittedCarrier": ['AA', 'UA', 'LA', 'DL']
          }         
        ],
            "passengers": {
            "adultCount": 1
        },
            "solutions": 100,
            "refundable": False
          }
        }
    
        r = requests.post(url, data=json.dumps(params), headers=headers)
        j = r.json()
        results.append(j)
        i += 1
    
    return results             
```

```python
sfoscl_dates = [dt.date(2015,4,12) + dt.timedelta(days=x) for x in range(0,7)]
sfoscl_dates2 = [dt.datetime.strftime(x, '%Y-%m-%d') for x in sfoscl_dates]

sfoscl = qfx(sfoscl_dates2, 'SFO', 'SCL')
```
